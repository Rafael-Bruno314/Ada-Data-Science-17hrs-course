```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel('GUIAS DE ITBI PAGAS (2024).xlsx')
df = df.drop(['N° do Cadastro (SQL)', 'Nome do Logradouro', 'Complemento', 'Referência', 'CEP', 'Situação do SQL', 'Área do Terreno (m2)', 'Testada (m)',	'Fração Ideal',	'Área Construída (m2)',	'Uso (IPTU)', 'Valor Venal de Referência',	'Proporção Transmitida (%)','Valor Venal de Referência (proporcional)','Base de Cálculo adotada','Tipo de Financiamento','Valor Financiado','Cartório de Registro','Matrícula do Imóvel','Descrição do uso (IPTU)','Padrão (IPTU)','Descrição do padrão (IPTU)','ACC (IPTU)'], axis=1)

df.head() 
```

```python
vendas_por_bairro = df['Bairro'].value_counts().reset_index()
vendas_por_bairro.columns = ['Bairro', 'Quantidade']

vendas_por_bairro_filtrado = vendas_por_bairro[vendas_por_bairro['Quantidade'] > 64]

vendas_por_bairro_filtrado
```

```python
# Plotar o gráfico de barras
plt.bar(vendas_por_bairro_filtrado['Bairro'], vendas_por_bairro_filtrado['Quantidade'], color='limegreen')

# Adicionar título e rótulos aos eixos
plt.title('Quantidade de Casas Vendidas por Bairro')
plt.xlabel('Bairro')
plt.ylabel('Quantidade de Casas Vendidas')

# Rotacionar os rótulos do eixo x para melhor visualização
plt.xticks(rotation=85)

# Mostrar o gráfico
plt.tight_layout()
plt.show()
```
