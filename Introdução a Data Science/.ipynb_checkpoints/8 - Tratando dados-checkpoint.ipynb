```python
def limpa_preco(linha):
    linha['Natureza de Transação'] = linha['Natureza de Transação'].replace('1.','')
    return linha
```

```python
df.apply(lambda x: limpa_preco(x), axis=1) #axis=1 é rodar pelas linhas (igual a 0 é pelas colunas)
```
